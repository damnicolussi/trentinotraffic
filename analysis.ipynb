```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import util.previous as vehicles
import seaborn as sns
sns.set(rc={'figure.figsize':(11, 4)})
```

```python
today = datetime.date.today()
```

## Vehicles per hour in a day

```python
hour = {}
hour['hour'] = [i for i in range(24)]
hour['veichles'] = vehicles.hour_vehicles
h_vehicles = pd.DataFrame(hour)

h_vehicles.plot.bar(x='hour', y='veichles', title=f'Veichles per hour ({today.strftime("%d/%m/%Y")})')
plt.show()
```

## Vehicles per day in a week

```python
start_of_week = today - datetime.timedelta(days=today.weekday())
end_of_week = start_of_week + datetime.timedelta(days=6)

day = {}
day['day'] = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
day['veichles'] = vehicles.day_vehicles
d_vehicles = pd.DataFrame(day)

d_vehicles.plot.bar(x='day', y='veichles', title=f'Veichles per day ({start_of_week.strftime("%d/%m/%Y")} - {end_of_week.strftime("%d/%m/%Y")})')
plt.show()
```

## Veichles per month in a year

```python
month = {}
month['month'] = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
month['veichles'] = vehicles.month_vehicles
m_vehicles = pd.DataFrame(month)

m_vehicles.plot.bar(x='month', y='veichles', title=f'Veichles per month ({today.strftime("%Y")})')
plt.show()
```
